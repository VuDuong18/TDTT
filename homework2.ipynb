```python
import math

# Nhập hệ số a, b, c
a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))

# Kiểm tra nếu a = 0 thì thành phương trình bậc 1
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        x = -c / b
        print("Phương trình bậc 1, nghiệm x =", x)
else:
    # Tính delta
    delta = b**2 - 4*a*c
    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có 2 nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép x =", x)
    else:
        print("Phương trình vô nghiệm thực")
```

```python
# Nhập hệ số
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
m = float(input("Nhập m: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
n = float(input("Nhập n: "))

# Tính định thức
D = a*d - b*c

if D != 0:
    # Hệ phương trình có nghiệm duy nhất
    Dx = m*d - b*n
    Dy = a*n - m*c
    x = Dx / D
    y = Dy / D
    print("Hệ phương trình có nghiệm duy nhất:")
    print("x =", x)
    print("y =", y)
else:
    # Kiểm tra trường hợp vô số nghiệm hay vô nghiệm
    if a*n - c*m == 0 and b*n - d*m == 0:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
```

```python
# Nhập số giây
x = int(input("Nhập thời gian (giây): "))

# Tính số giờ, phút, giây
gio = x // 3600              # 1 giờ = 3600 giây
phut = (x % 3600) // 60      # Lấy phần còn lại sau khi trừ giờ, chia 60 để ra phút
giay = x % 60                 # Phần còn lại là giây

# In kết quả
print(gio, "giờ", phut, "phút", giay, "giây")
```

```python
# Nhập tâm và bán kính đường tròn
h = float(input("Nhập hoành độ tâm (h): "))
k = float(input("Nhập tung độ tâm (k): "))
r = float(input("Nhập bán kính r: "))

# Nhập điểm A(x, y)
x = float(input("Nhập hoành độ điểm A (x): "))
y = float(input("Nhập tung độ điểm A (y): "))

# Tính khoảng cách từ điểm A đến tâm
khoang_cach = ((x - h)**2 + (y - k)**2)**0.5

# Kiểm tra
if khoang_cach == r:
    print("Điểm A nằm trên đường tròn")
else:
    print("Điểm A không nằm trên đường tròn")
```
