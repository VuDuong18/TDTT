```python
# Bài 2
# Nhập dữ liệu
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))

# Hằng số pi
pi = 3.14

# Tính bán kính hình tròn (nằm ở trung tâm khu đất)
r = min(a, b) / 2

# Tính diện tích
S_chunhat = a * b
S_tron = pi * (r ** 2)

# Diện tích còn lại để trồng cây
S_trongcay = S_chunhat - S_tron

# In kết quả, làm tròn 2 chữ số thập phân
print("Diện tích trồng cây là:", round(S_trongcay, 2))
```

```python
# Nhập dữ liệu: 6 điểm trên cùng 1 dòng, cách nhau bởi dấu cách
data = input("Nhập 6 điểm a1 b1 c1 a2 b2 a3: ")

# Chuyển dữ liệu thành danh sách các số thực
a1, b1, c1, a2, b2, a3 = map(float, data.split())

# Tính điểm trung bình theo công thức
TB = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10

# In kết quả, làm tròn 1 chữ số thập phân
print("Điểm trung bình:", round(TB, 1))

```

```python
# Nhập vào một kí tự
c = input("Nhập một ký tự: ")

# Kiểm tra và chuyển đổi chữ hoa <-> chữ thường
if c.isupper():        # Nếu là chữ hoa
    print(c.lower())    # In chữ thường tương ứng
else:                   # Ngược lại (chữ thường)
    print(c.upper())    # In chữ hoa tương ứng
```

```python
# Nhập vào một kí tự
c = input("Nhập một ký tự: ")

# Kiểm tra xem c có phải là chữ cái không
if c.isalpha():
    print(f"{c} là kí tự alphabet")
else:
    print(f"{c} không phải là kí tự alphabet")
```

```python
# Nhập vào một chữ cái hoa
c = input("Nhập một chữ cái hoa: ")

# Chuyển chữ hoa thành chữ thường tương ứng
lower_c = c.lower()

# Xét trường hợp đặc biệt khi c = 'A'
if c == 'A':
    print("Trường hợp đặc biệt: trước 'a' không có chữ cái nào")
else:
    # Tính chữ cái liền trước chữ thường tương ứng
    prev_char = chr(ord(lower_c) - 1)
    print("Chữ cái liền trước của", lower_c, "là", prev_char)
```
