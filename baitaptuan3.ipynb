```python
a = 5   # 101 (bit)
b = 3   # 011 (bit)

print("a =", a, "->", bin(a))
print("b =", b, "->", bin(b))

print("\n--- Phép toán BIT ---")

# AND
print("a & b =", a & b, "->", bin(a & b))

# OR
print("a | b =", a | b, "->", bin(a | b))

# XOR
print("a ^ b =", a ^ b, "->", bin(a ^ b))
```

```python
def is_power_of_two(n):
    if n <= 0:
        return False
    return (n & (n - 1)) == 0

# Thử nghiệm
numbers = [1, 2, 3, 4, 8, 12, 16, 0, -4]

for num in numbers:
    print(num, "->", is_power_of_two(num))
```

```python
# Nhập số thực
x = float(input("Nhập một số thực: "))

# Tách phần nguyên và phần thập phân
if x >= 0:
    phan_nguyen = int(x)
    phan_thap_phan = x - phan_nguyen
else:
    phan_nguyen = int(x)
    phan_thap_phan = x - phan_nguyen

# Làm tròn xuống
if x >= 0:
    floor = phan_nguyen
else:
    floor = phan_nguyen if phan_thap_phan == 0 else phan_nguyen - 1

# Làm tròn lên
if x >= 0:
    ceil = phan_nguyen if phan_thap_phan == 0 else phan_nguyen + 1
else:
    ceil = phan_nguyen

# Làm tròn tới số nguyên gần nhất
if abs(phan_thap_phan) >= 0.5:
    if x >= 0:
        round_nearest = phan_nguyen + 1
    else:
        round_nearest = phan_nguyen - 1
else:
    round_nearest = phan_nguyen

# In kết quả cách nhau một dấu cách
print(f"{ceil} {floor} {round_nearest}")
```
